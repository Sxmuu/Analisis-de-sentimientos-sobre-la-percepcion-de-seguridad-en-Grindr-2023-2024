```python
df = pd.read_csv('google_play_comments_all_countries.csv')

df.head()
```

```python
df.drop(columns=['texto_Traducido'], inplace=True)

df.head()
```

```python
###NLTK
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
as1 = []
```

```python
contenido = df["content"]
contenido_lista = contenido.values.tolist()
```

```python
for i in contenido_lista[:11000]:
  b = sia.polarity_scores(i)
  as1.append(b)
```

```python
bd_2 = {"Comentario": contenido_lista, "Fecha": df['at'], "Polaridad":as1}
df_2 = pd.DataFrame(bd_2)
df_2
```
