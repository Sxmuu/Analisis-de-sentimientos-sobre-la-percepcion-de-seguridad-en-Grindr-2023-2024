```python
from google_play_scraper import reviews
import pandas as pd

# ID de la aplicación Nequi
app_id = 'com.ubercab'

# Especificar el idioma y el país
lang = 'es'      # Español
country = 'co'   # Colombia

# Obtener comentarios con filtros de idioma y país
result, continuation_token = reviews(
    app_id,
    lang=lang,
    country=country,
    count=1000,        # Cantidad inicial de comentarios
    filter_score_with=None  # Sin filtro de calificación
)

# Convertir el resultado inicial en un DataFrame
commentsDF = pd.DataFrame(result)

# Definir el límite máximo de iteraciones para obtener más comentarios
max_iterations = 10
iterations = 0

# Obtener más comentarios usando el continuation_token
while continuation_token and iterations < max_iterations:
    more_results, continuation_token = reviews(
        app_id,
        continuation_token=continuation_token,
        lang=lang,
        country=country,
        count=1000,
        filter_score_with=None
    )

    # Convertir los nuevos resultados en un DataFrame y concatenarlos con el DataFrame original
    more_commentsDF = pd.DataFrame(more_results)
    commentsDF = pd.concat([commentsDF, more_commentsDF], ignore_index=True)

    # Incrementar el contador de iteraciones
    iterations += 1

# Mostrar una muestra aleatoria de 10 comentarios
print(commentsDF.sample(10))
```

```python
import pandas as pd
from googletrans import Translator

# Cargar el archivo CSV
file_path = 'google_play_comments_all_countries.csv'
df = pd.read_csv(file_path)

# Crear una instancia del traductor
translator = Translator()

# Función para traducir el texto a inglés, dejando el texto en inglés sin cambios
def translate_to_english(text):
    try:
        detected_lang = translator.detect(text).lang
        if detected_lang != 'en':
            translated = translator.translate(text, src=detected_lang, dest='en')
            return translated.text
        else:
            return text
    except Exception as e:
        print(f"Error translating text: {text}\n{e}")
        return text

# Aplicar la función de traducción a la columna de comentarios
df['texto_Traducido'] = df['content'].apply(translate_to_english)

# Guardar el DataFrame con los comentarios traducidos en un archivo CSV
output_file_path = 'google_play_comments_translated.csv'
df.to_csv(output_file_path, index=False)

# Mostrar las primeras filas del DataFrame con la nueva columna de comentarios traducidos
print(df.head())
```

```python
import pandas as pd

# Cargar el archivo CSV
file_path = 'path_to_your_file/google_play_comments_all_countries.csv'
df = pd.read_csv(file_path)

# Función para verificar si una fila contiene alguna de las palabras clave
def contains_keywords(text, keywords):
    for keyword in keywords:
        if keyword.lower() in text.lower():
            return True
    return False

# Filtrar las filas que contienen alguna de las palabras clave
df_filtered = df[df['content'].apply(lambda x: contains_keywords(x, keywords))]

# Guardar el DataFrame filtrado en un archivo CSV
output_file_path = 'google_play_comments_transformed_and_filtered.csv'
df_filtered.to_csv(output_file_path, index=False)

# Mostrar las primeras filas del DataFrame filtrado
df_filtered.head()
```
